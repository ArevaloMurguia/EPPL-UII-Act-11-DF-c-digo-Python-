```python
# Calcular el promedio de 4 calificaciones
# Ingresar los valores de C1, C2, C3, C4
# Formula para calcular el promedio
# Promedio = (C1 + C2 + C3 + C4) / 4
# Mostrar el valor del promedio

C1 = float(input("Ingrese la calificacion 1: "))
C2 = float(input("Ingrese la calificacion 2: "))
C3 = float(input("Ingrese la calificacion 3: "))
C4 = float(input("Ingrese la calificacion 4: "))

S = C1 + C2 + C3 + C4
P = S / 4

print("El promedio es:", P)
print("Programa realizado por Arevalo Rafael 0031")
```
